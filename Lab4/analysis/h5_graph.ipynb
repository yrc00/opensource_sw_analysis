```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
df = pd.read_csv("modified_h5_summary.csv")

# 스타일 설정
sns.set(style="whitegrid")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# === 시각화 스타일 설정 ===
sns.set(style="whitegrid", palette="muted", font_scale=1.2)
```

```python
df = df.groupby(['alpha', 'preclude', 'preserve']).mean().reset_index()
```

## Write

### Time

```python
df[['alpha', 'case', 'time(s)', 'compaction_time(s)', 'stall_flush_total(s)']]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# alpha 값 목록 추출
alphas = sorted(df['alpha'].unique())

# subplot 설정
num_plots = len(alphas)
fig, axes = plt.subplots(nrows=num_plots, figsize=(14, 5 * num_plots), squeeze=False)

# 각 alpha에 대해 subplot 생성
for idx, alpha_val in enumerate(alphas):
    ax = axes[idx][0]

    # 해당 alpha 값에 대한 데이터 추출
    subset = df[df['alpha'] == alpha_val].copy()

    # case를 문자열로 변환하여 x축 index로 설정
    subset['case_str'] = subset['case'].astype(str)
    plot_df = subset.set_index('case_str')[['time(s)', 'compaction_time(s)', 'stall_flush_total(s)']]

    # bar plot
    bars = plot_df.plot(kind='bar', ax=ax)
    ax.set_title(f"Write Workload: Time Comparison at alpha = {alpha_val}")
    ax.set_ylabel("Time (seconds)")
    ax.set_xlabel("case")
    ax.set_xticklabels(plot_df.index, rotation=45, ha='right')

    # 각 막대 위에 값 표시
    for container in bars.containers:
        bars.bar_label(container, fmt='%.2f', padding=3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# case가 문자열이면 숫자형으로 변환 (필요시)
df['case'] = df['case'].astype(int)

# 꺾은선 그래프 그리기: x=case, y=time(s), alpha별로 선 구분
sns.lineplot(data=df, x='case', y='time(s)', hue='alpha', marker='o')

plt.title('Write Workload: Time(s) by case for different alpha values')
plt.xlabel('case')
plt.ylabel('time (seconds)')
plt.xticks(sorted(df['case'].unique()))
plt.legend(title='alpha')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### WAF

```python
df[['alpha', 'case', 'WAF']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# alpha 값 목록 추출 및 정렬
alphas = sorted(df['alpha'].unique())

num_plots = len(alphas)
fig, axes = plt.subplots(nrows=num_plots, figsize=(14, 5 * num_plots), squeeze=False)

for idx, alpha_val in enumerate(alphas):
    ax = axes[idx][0]
    
    subset = df[df['alpha'] == alpha_val]
    
    # x축 레이블을 'case' 컬럼 값으로 설정 (문자열로 변환)
    x_labels = subset['case'].astype(str)
    
    sns.barplot(x=x_labels, y=subset['WAF'], ax=ax)
    ax.set_title(f"Write Amplification Factor (WAF) by case at alpha={alpha_val}")
    ax.set_xlabel("case")
    ax.set_ylabel("WAF")
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    
    # 막대 위에 WAF 값 표시
    for p in ax.patches:
        height = p.get_height()
        ax.text(
            p.get_x() + p.get_width() / 2,
            height * 1.01,
            f'{height:.3f}',
            ha='center',
            va='bottom',
            fontsize=9
        )

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# case가 문자열이면 숫자형으로 변환 (필요 시)
df['case'] = df['case'].astype(int)

plt.figure(figsize=(12, 6))

# 꺾은선 그래프 그리기, x=case, y=WAF, alpha별로 색 구분
sns.lineplot(data=df, x='case', y='WAF', hue='alpha', marker='o')

plt.title('Write Amplification Factor (WAF) by case for different alpha values')
plt.xlabel('case')
plt.ylabel('WAF')
plt.xticks(df['case'].unique())  # case별로 x축 눈금 설정
plt.legend(title='alpha')

plt.grid(True)
plt.tight_layout()
plt.show()
```
