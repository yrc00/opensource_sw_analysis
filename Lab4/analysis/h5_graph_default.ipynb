```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)  # 모든 열 표시
pd.set_option('display.max_rows', None)  # 모든 행 표시

# 데이터 로드
df_1 = pd.read_csv("modified_h5_default_summary_with_case.csv")
df_1
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
df_2 = pd.read_csv("modified_h5_summary.csv")
df_2 = df_2[(df_2['preclude'] == 0) & (df_2['preserve'] == 0)]
df_2
```

```python
# df_1, df_2에 type 컬럼 추가
df_1['type'] = 'default'
df_2['type'] = 'tiered'

# 두 데이터프레임 합치기
df = pd.concat([df_1, df_2], ignore_index=True)
df
```

### DBSize

```python
df_1[['alpha', 'case', 'db_size_bytes']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# seaborn 스타일 설정
sns.set(style="whitegrid")

# 그래프 크기 설정
plt.figure(figsize=(10, 6))

# barplot: case_str을 기준으로 x축, hue=alpha
sns.barplot(data=df, x='alpha', y='db_size_bytes', hue='type')

# 축 및 제목 설정
plt.title('DB Size by Case and Alpha')
plt.xlabel('Case')
plt.ylabel('DB Size (bytes)')

# 막대 위에 값 표시
for container in plt.gca().containers:
    plt.bar_label(container, fmt='%.2e', label_type='edge', fontsize=8)

plt.tight_layout()
plt.legend(title='Alpha')
plt.show()
```
